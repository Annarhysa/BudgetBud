```python
import pandas as pd
import numpy as np
import os
```

```python
# Specify the path of the directory you want to open
directory_path = 'C:/Users/annar/OneDrive/Desktop/Visual Studio/GitHub/Hackathons/BudgetBud/backend/data'

# Change the current working directory to the specified directory
os.chdir(directory_path)

data = pd.read_csv("expense.csv")
data.head()
```

```python
data['category'] = data['category'].replace('Sourav ', 'Miscellaneous')
data['category'] = data['category'].replace('Sourav', 'Miscellaneous')
data['category'] = data['category'].replace('Eggs', 'Groceries')
data['category'] = data['category'].replace('Friends ', 'Friends')
data['category'].unique()
```

```python
# Remove the "-" sign from the column
data['amount'] = data['amount'].str.replace(',', '')
data['amount'] = pd.to_numeric(data['amount'], errors='coerce')
data['amount'] = data['amount'].abs()
data.head()
```

```python
#preprocessing the data
columns = ['converted amount', 'currency', 'currency.1']
data.drop(columns=columns, inplace=True)
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
data.head()
```

```python
data = data.sort_values(by = 'date')
data.head()
```

```python
# Extract month from the 'Date' column
data['Month'] = data['date'].dt.month
data['Month'] = data['Month'].replace(10, 'October')
data['Month'] = data['Month'].replace(11, 'November')
data['Month'] = data['Month'].replace(12, 'December')
data['Month'] = data['Month'].replace(1, 'January')
data['Month'] = data['Month'].replace(2, 'February')
data['Month'] = data['Month'].replace(3, 'March')
data.head()
```

```python
#converting preprocessed data to csv
data.to_csv('data.csv')
```

```python
#calculating the sum of monthly expenses
monthly_expenses = data.groupby('Month')['amount'].sum()
monthly_expenses = monthly_expenses.to_dict()
monthly_expenses
```

```python
# Create a DataFrame from the dictionary
expenses_df = pd.DataFrame(list(monthly_expenses.items()), columns=['Month', 'MonthlyExpenses'])
expenses_df.head()
```

```python
#calculating monthly expenditure according to categories
categorical_expenses = data.groupby(['category', 'Month'])['amount'].sum()
monthly_sum_df = categorical_expenses.reset_index().pivot_table(index='Month', columns='category', values='amount')
```

```python
# Arrange months in ascending order
custom_order = ['October', 'November', 'December', 'January', 'February', 'March']
monthly_sum_df = monthly_sum_df.fillna(0)
monthly_sum_df.reindex(custom_order)
monthly_sum_df.head()
```

```python
#need to add the monthly expenses
merged_df = pd.merge(monthly_sum_df, expenses_df, on='Month', how='left')
merged_df
```

```python
# Convert the 'Name' column to categorical data type with the desired order
merged_df['Month'] = merged_df['Month'].astype(pd.CategoricalDtype(custom_order, ordered=True))

# Sort the DataFrame based on the 'Name' column
df_sorted = merged_df.sort_values('Month')
df_sorted.reset_index(drop=True, inplace=True)
df_sorted
```

```python
#converting this to a new csv
df_sorted.to_csv('monthly_data.csv')
```
